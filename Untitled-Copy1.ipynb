```python
import numpy as np
required_urls = []
for word in ['forex (exchange AND rate)','(federal AND reserve) (federal AND funds AND rate ) ','fed federal (federal AND rate)', 'inflation','(rates AND interest)','bonds','(monetary AND policy)', '(central AND bank)', '(quantitative AND easing)'  ]:
    
    required_urls.append(list({v['id']:v for v in list(np.load('url7'+word +'.npy'))}.values()))
    
    print(word, len(required_urls[-1]), required_urls[-1][-1])
    print()
```

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
from dateutil.parser import parse
from tqdm import  tqdm_notebook as tqdm
def getTextFromURL(url):
    info = {}
    try : 
        html = urlopen(url).read()
        soup = BeautifulSoup(html, "lxml")
        #title = soup.title.string
        article_text = ''
        try : 
            article = soup.find("div", {"class":"content__article-body from-content-api js-article__body"}).findAll('p')
            for element in article:
                article_text += ''.join(element.findAll(text = True))
        except AttributeError : 
            article = soup.findAll("div", {"class":"block-elements block-elements--no-byline"})
            for i in range(len(article)):
                for element in article[i].findAll('p'):
                            article_text += ''.join(element.findAll(text = True))
                
        
    except HTTPError :
        print(url)
        return('')
    return(article_text)
def getDescriptionFromURL(url):
    info = {}
    try : 
        html = urlopen(url).read()
        soup = BeautifulSoup(html, "lxml")
        try : 
            description = soup.find(itemprop="description").get("content")
        except AttributeError:
            description = ''
    except HTTPError :
        print(url)
        return('')        
    return(description)
```

```python
list_df = []
```

```python
import pandas as pd
from urllib.error import HTTPError
for i in range(len(required_urls)-1,3,-1):
    r = required_urls[i]
    
    df = pd.DataFrame()
    df['date'] = list(map(lambda x : x['webPublicationDate'],r))
    
    df['sectionId']  = list(map(lambda x : x['sectionId'],r))
    df['sectionName']  = list(map(lambda x : x['sectionName'],r))
    df['title']  = list(map(lambda x : x['webTitle'],r))
    print('done')
    df['description']  = list(map(lambda x :  getDescriptionFromURL(x['webUrl']),r))
    print('done')

    df['text']= list(map(lambda x : getTextFromURL(x['webUrl']),r))
    list_df.append(df)
```

```python
np.save('liste.npy',list_df)
```
